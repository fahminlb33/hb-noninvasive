```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def pwm_map(x: float, in_min: float, in_max: float, out_min: float, out_max: float):
    return (x - in_min) * (out_max - out_min) / (in_max - in_min) + out_min
```

# Signal 1

```python
pwm_1_values = []
pwm_2_values = []
pwm_all_values = []

for i in range(3):
    for j in range(10):
        pwm_value = pwm_map(j, 0, 9, 0, 255)

        if i == 0:
            pwm_1_values.extend([pwm_value] * 10)
            pwm_2_values.extend([0] * 10)
            pwm_all_values.extend([0] * 10)
        elif i == 1:
            pwm_1_values.extend([0] * 10)
            pwm_2_values.extend([pwm_value] * 10)
            pwm_all_values.extend([0] * 10)
        elif i == 2:
            pwm_1_values.extend([0] * 10)
            pwm_2_values.extend([0] * 10)
            pwm_all_values.extend([pwm_value] * 10)

        # to simulate delay between measurements
        pwm_1_values.extend([0] * 10)
        pwm_2_values.extend([0] * 10)
        pwm_all_values.extend([0] * 10)
```

```python
x = np.linspace(0, 20, len(pwm_1_values))
plt.plot(x, pwm_1_values)
plt.plot(x, pwm_2_values)
plt.plot(x, pwm_all_values)
plt.xlabel("Durasi rekaman (s)")
plt.ylabel("Intensitas cahaya LED")
```
